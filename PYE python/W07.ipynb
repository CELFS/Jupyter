### 文件字符分布

```python
def getText():
    txt = open('dataSet\\latex.log', "r", encoding='utf-8').read()
    for ch in '0123456789!"#$%&()*+,-./:;<=>?@[\\]^_‘{|}~`\x14\x02\x1e\n':
        txt = txt.replace(ch, ' ')
    for ch in txt:
        if 'A' <= ch <= 'Z':
            txt = txt.replace(ch, ' ')
    return txt

s = getText()
cnts = {}
letters = [chr(i) for i in range(97, 123)]

for ch in s:
    if ch in letters: # if ch.islower()
        cnts[ch] = cnts.get(ch, 0) + 1
    else:
        continue

result = []

for i in cnts:
    tempStr = i + ':' + str(cnts.get(i))
    result.append(tempStr)

result.sort()
last = ""
print("共{}字符,".format(len(s)), end='')
for i in result:
    last = last + i + ','

print(last[:-1])
```

### 文件独特行数

* 统计附件文件中与其他任何其他行都不同的行的数量，即独特行的数量。
    * 独特行：没有重复的行，但凡重复了一次，就不叫独特行

```python
# with open(r'dataSet/latex.log', 'r') as f: # with 与直接 open 有什么不同？

f = open(r'dataSet/latex.log', 'r')

s = f.readlines()
s = set(s)
cnt = 0
for i in s:
    if i != '\n':
#         print(i)
        cnt += 1
print("共{}独特行".format(cnt))
# print(cnt)

f.close()
```

```python
# 参考答案
f = open(r'dataSet/latex.log', 'r')
ls = f.readlines()
s = set(ls)
# print(s)
for i in s:
#     ls.remove(i) # 没看懂这一步，去除独特行？
    print(i)
t = set(ls)
print("共{}独特行".format(len(s)-len(t))) # 最后把总的减去移除独特行的，就是独特行数量

f.close()
```

* 记住：如果需要"去重"功能，请使用集合类型。
* ls.remove()可以去掉某一个元素，如果该行是独特行，去掉该元素后将不在集合t中出现。

```python
# 独特行的测试
ls = [1, 2, 2, 3, 3, 3, 4]
s = set(ls) # {1, 2, 3, 4}

print("{0:4}{1}".format("ls:", ls))
print("{0:4}{1}".format("s:", s))

for i in s:
    ls.remove(i) # [2, 3, 3] 全部移除一次，剩下元素副本
    
t = set(ls) # {2, 3} 表示可产生副本的元素个数

print("{0:4}{1}".format("ls:", ls))
print("{0:4}{1}".format("t:", t))

print("{0:13}{2}\n{1:10}{3}".format("这是总体元素", "这是出现重复的元素", s, t))
print(len(s) - len(t)) # 总体去重个数 - 可产生副本的元素个数
```

### CSV 格式列变换

* 附件是一个CSV文件，请将每行按照列逆序排列后输出，不改变各元素格式（如【周围空格】布局等）。

```python
with open(r'dataSet/data.csv', 'r') as f:
    for line in f.readlines():
        line = line.strip("\n")
#         s = line[::-1] # 直接不可行，会改变空格
#         print(s)
        ls = line.split(",") # 需按逗号分割数据，元素内的空格关系保留
#         print(ls)
        ls = ls[::-1]
        print(",".join(ls))
        
f.close()
```
